```python
# s3 11899 괄호 끼워넣기
import sys
input = sys.stdin.readline
res = stack = 0

for c in input():
    if c == ")":
        res += not bool(stack)
        stack -= bool(stack)
    elif c == "(":
        stack += 1
print(res + stack)

# s2 17827 달팽이 리스트
import sys
input = sys.stdin.readline

N, M, V = map(int, input().split())
lst = list(map(int, input().split()))

for _ in range(M):
    k = int(input())
    
    if k < V:
        print(lst[k])
    else:
        print(lst[(k-V+1)%(N-V+1) + V - 1])

# s5 27111번 출입기록
import sys
input = sys.stdin.readline

inside = set()
error_cnt = 0

for _ in range(int(input())):
    a, b = map(int, input().split())
    if b:
        if a in inside:
            error_cnt += 1
            continue
        inside.add(a)
    else:
        if a not in inside:
            error_cnt += 1
            continue
        inside.discard(a)

print(error_cnt + len(inside))
```

```python
# s1 15966번 군계일학 # https://www.acmicpc.net/problem/15966 # 미해결
import sys
input = sys.stdin.readline

N = int(input())
arr = list(map(int, input().split()))
dic = {}

for i in range(N):
    dic[arr[i] - i - 1] = dic.get(arr[i] - i - 1, 0) + 1

print(max(dic.values()))

# s1 18243번 Small World Network # https://www.acmicpc.net/problem/18243
# 그래프이론 # 최단경로 # 플로이드 워셜
import sys
input = sys.stdin.readline

N, K = map(int, input().split())

graph = [[float("inf")]*N for _ in range(N)]

for i in range(N):
    graph[i][i] = 0
for _ in range(K):
    s, e = map(int, input().split())
    graph[s-1][e-1] = 1
    graph[e-1][s-1] = 1

for k in range(N):
    for s in range(N):
        for e in range(N):
            graph[s][e] = min(graph[s][e], graph[s][k] + graph[k][e])

if all(graph[i][j] <= 6 for i in range(N) for j in range(N)):
    print("Small World!")
else:
    print("Big World!")

# s4 16162번 가희와 3단 고음 # https://www.acmicpc.net/problem/16162
import sys
input = sys.stdin.readline

N, A, D = map(int, input().split())
sound = list(map(int, input().split()))
cnt = 0

for n in sound:
    if A == n:
        A += D
        cnt += 1

print(cnt)
```

# 3주차

```python
# s2 1326번 폴짝폴짝 https://www.acmicpc.net/problem/1326
# BFS # 풀이시간 초과해결
import sys
from collections import deque
input = sys.stdin.readline

N = int(input())
adj_lst = [[] for _ in range(N+1)]

for i, d in enumerate(map(int, input().split()), 1):
    for n in range(i%d, i, d):
        adj_lst[i].append(n)
    for n in range(i+d, N+1, d):
        adj_lst[i].append(n)

start, target = map(int, input().split())
visit = deque([start])
visited = [-1]*(N+1)
visited[start] = 0

while visit:
    cur = visit.popleft()
    
    if cur == target:
        print(visited[cur])
        break
    
    for nxt in adj_lst[cur]:
        if visited[nxt] == -1:
            visited[nxt] = visited[cur] + 1
            visit.append(nxt)
else:
    print(-1)

# g5 10597번 순열장난 https://www.acmicpc.net/problem/10597
# 미해결
import sys
input = sys.stdin.readline

permutation = input()

# s2 14248번 점프 점프 https://www.acmicpc.net/problem/14248
# BFS&DFS
import sys
input = sys.stdin.readline

N = int(input())
adj_lst = [[] for _ in range(N+1)]

for i, d in enumerate(map(int, input().split()), 1):
    if i+d <= N:
        adj_lst[i].append(i+d)
    if i-d >= 1:
        adj_lst[i].append(i-d)

visit = [int(input())]
visited = set()

while visit:
    cur = visit.pop()
    visited.add(cur)
    
    for nxt in adj_lst[cur]:
        if nxt not in visited:
            visit.append(nxt)

print(len(visited))
```

```python
# 쿠키의 신체 측정 https://www.acmicpc.net/problem/20125 s4
import sys
input = sys.stdin.readline

board = [input() for _ in range(int(input()))]

for i, r in enumerate(board):
    for j, c in enumerate(r):
        if c == "*":
            heart = (i+2, j+1)
            break
    else:
        continue
    break

arm1 = j - board[i+1].find("*")
arm2 = board[i+1].rfind("*") - j
waist = 1

while board[i+3][j] == "*":
    i += 1
    waist += 1

leg1 = i
leg2 = i

for r in range(i+3, len(board)):
    if board[r][j-1] == "*":
        leg1 = r + 1 - heart[0] - waist
    if board[r][j+1] == "*":
        leg2 = r + 1 - heart[0] - waist

print(*heart)
print(arm1, arm2, waist, leg1, leg2)

# 문자열 교환 https://www.acmicpc.net/problem/1522 s1
def string_exchange(string):
    a_n = string.count("a")
    if a_n == 0 | len(string):
        return 0
    string += string
    b_n = string[:a_n].count("b")
    res = b_n
    
    for i in range(len(string)//2 - 1):
        b_n += (string[i+a_n]=="b") - (string[i]=="b")
        res = min(res , b_n)
    
    return res

string = input()
print(string_exchange(string))

# 색칠 1 https://www.acmicpc.net/problem/1117 g5
def coloring(W, H, f, c, x1, y1, x2, y2):
    return W*H - (c + 1)*(y2 - y1)*(x2-x1 + max(0, min(f, W-f, x2) - x1))

W, H, f, c, x1, y1, x2, y2 = map(int, input().split())
print(coloring(W, H, f, c, x1, y1, x2, y2))
```
